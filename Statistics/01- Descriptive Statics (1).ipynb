```python
# Importing useful Python libraries

import numpy as np              # numpy → used for numbers, arrays, and math operations
import seaborn as sns           # seaborn → used for creating beautiful graphs
import matplotlib.pyplot as plt # matplotlib → used for plotting graphs
import pandas as pd             # pandas → used for handling and analyzing data (tables, CSV, Excel etc.)
```

```python
# Reading a CSV (Excel-like file) into a pandas DataFrame
# df → will store the data in table format
df = pd.read_csv(r"C:\Users\hp\OneDrive\Pictures\Desktop\FSDS,Agentic AI\SEPTEMBER\EXCEL Sheets\Inc_Exp_Data.csv")
```

```python
# Just writing df will display the DataFrame (table) contents
# It shows the data we loaded from the CSV file
df
```

```python
# Display information about the DataFrame
# It shows:
# - total number of rows
# - column names
# - number of non-empty values in each column
# - data type of each column (like int, float, object/string)
df.info()
```

```python
# Show basic statistics of numeric columns in the DataFrame
# df.describe() gives:
# - count → number of values
# - mean → average value
# - std → standard deviation (how spread out the numbers are)
# - min → smallest value
# - 25%, 50%, 75% → quartiles (divide data into 4 parts)
# - max → largest value
# .T → transpose the result to make rows as columns (easier to read)
df.describe().T
```

```python
# Show the shape of the DataFrame
# It returns a tuple: (number of rows, number of columns)
df.shape
```

```python
# Check for missing values in each column
# df.isnull() → returns True for missing values, False otherwise
# .sum() → counts how many missing values are in each column
df.isnull().sum()
```

```python
# Check if any column has at least one missing value
# df.isna() → True for missing values, False otherwise (similar to isnull)
# .any() → returns True if there is at least one missing value in the column
df.isna().any()
```

```python
# Calculate the average (mean) of the 'Mthly_HH_Expense' column
# df['Mthly_HH_Expense'] → selects the column
# .mean() → calculates the average value of that column
df['Mthly_HH_Expense'].mean()
```

```python
# Calculate the median of the 'Mthly_HH_Expense' column
# Median → the middle value when all numbers are sorted
# If there are even numbers of values, it gives the average of the two middle values
df['Mthly_HH_Expense'].median()
```

```python
# Create a frequency table of 'Mthly_HH_Expense'
# pd.crosstab() counts how many times each value appears
# index=df['Mthly_HH_Expense'] → rows will be the unique expense values
# columns='count' → just a label for the counts
mth_exp_tmp = pd.crosstab(index=df['Mthly_HH_Expense'], columns='count')

# Reset the index so that 'Mthly_HH_Expense' becomes a normal column again
mth_exp_tmp.reset_index(inplace=True)

# Print the frequency table
print(mth_exp_tmp)
```

```python
import matplotlib.pyplot as plt  # Importing matplotlib for plotting graphs

# Plot a bar chart of the 'Highest_Qualified_Member' column
# .value_counts() → counts how many times each qualification appears
# .plot(kind='bar') → creates a bar chart
# color='skyblue' → fill color of bars
# edgecolor='black' → border color of bars
df['Highest_Qualified_Member'].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')

# Set the title of the plot
plt.title("Qualification Distribution of Household Members")

# Set label for x-axis
plt.xlabel("Qualification")

# Set label for y-axis
plt.ylabel("Count")

# Display the plot
plt.show()
```

```python
# Plot a simple line/scatter chart
# x → 'Mthly_HH_Income' on X-axis
# y → 'Mthly_HH_Expense' on Y-axis
df.plot(x='Mthly_HH_Income', y='Mthly_HH_Expense')

# Calculate the Interquartile Range (IQR) for 'Mthly_HH_Expense'
# IQR → difference between 75th percentile (Q3) and 25th percentile (Q1)
# It shows the range of the middle 50% of the data
IQR = df['Mthly_HH_Expense'].quantile(0.75) - df['Mthly_HH_Expense'].quantile(0.25)

# Display the IQR value
IQR
```

```python
# Plot a simple line/scatter chart
# x → 'Mthly_HH_Income' on X-axis
# y → 'Mthly_HH_Expense' on Y-axis
df.plot(x='Mthly_HH_Income', y='Mthly_HH_Expense')

# Calculate a modified IQR using median instead of Q3
# IQR here = 50th percentile (median) - 25th percentile (Q1)
# It shows the range from Q1 to the median
IQR = df['Mthly_HH_Expense'].quantile(0.50) - df['Mthly_HH_Expense'].quantile(0.25)

# Display the IQR value
IQR
```

```python
# Calculate standard deviation for the first 5 columns of the DataFrame
# df.iloc[:, 0:5] → selects all rows and first 5 columns
# .std() → computes standard deviation of each column
# .to_frame() → converts the result to a DataFrame
# pd.DataFrame(...).T → makes it a DataFrame and then transpose it so columns become rows
pd.DataFrame(df.iloc[:, 0:5].std().to_frame()).T
```

```python
# Calculate variance for the first 4 columns of the DataFrame
# df.iloc[:, 0:4] → selects all rows and first 4 columns
# .var() → computes variance of each column (measure of how spread out the data is)
# .to_frame() → converts the result into a DataFrame
# pd.DataFrame(...).T → makes it a DataFrame and then transpose it so columns become rows
pd.DataFrame(df.iloc[:, 0:4].var().to_frame()).T
```

```python
# Count how many times each qualification appears in 'Highest_Qualified_Member' column
# .value_counts() → counts the occurrences of each unique value
# .to_frame() → converts the counts into a DataFrame
# .T → transpose the DataFrame so that qualifications become columns
df['Highest_Qualified_Member'].value_counts().to_frame().T
```

```python
# Plot a bar chart of the 'No_of_Earning_Members' column
# .value_counts() → counts how many households have each number of earning members
# .plot(kind='bar') → creates a vertical bar chart
df['No_of_Earning_Members'].value_counts().plot(kind='bar')
```

```python
# Create a box plot for the 'No_of_Earning_Members' column
# sns.boxplot() → draws a box plot to show distribution of data
# data=df → use this DataFrame
# x='No_of_Earning_Members' → column to plot
df = sns.boxplot(data=df, x='No_of_Earning_Members')
```
