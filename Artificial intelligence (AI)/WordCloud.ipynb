```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Sample text
text = """
Python is an amazing programming language.
Moiz loves learning Python, data science, AI, and machine learning.
Libraries like NumPy, Pandas, Seaborn, and Matplotlib make Python powerful!
"""

# Create the word cloud
wordcloud = WordCloud(
    width=900,
    height=500,
    background_color='black',
    colormap='plasma',      # change color theme
    contour_color='white',  # white outline
    contour_width=2,
    max_words=100,
    collocations=False      # avoid repeating word pairs
).generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Moiz's Python Learning Word Cloud", fontsize=16, color='white')
plt.show()
```
