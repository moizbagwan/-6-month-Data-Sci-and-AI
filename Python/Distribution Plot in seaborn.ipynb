```python
import seaborn as sns
import matplotlib.pyplot as plt
```

#We ue distribution plot to visualize distribution of Quantitative data.

    Lets discuss some plots thats allows visualization the distribution of data set

- Distplot/Histplot
- Joinplot
- Pairplot
- Rugplot

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique
```

```python
sns.histplot(df['total_bill'])
```

```python
sns.histplot(df['total_bill'],bins=20)
```

```python
sns.histplot(df['total_bill'],bins=20,kde=True)
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=20,kde=True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

```python
sns.histplot(df['total_bill'],color="Red")
```

# Jointplot

    - sns.jointplot shows the relationship between two variables along with their individual distributions.

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = df,kind="scatter")
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = df,kind="hex")
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = df,kind="reg")
```

# Pairplot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='size',palette='rainbow')
```

# Rugplot

```python
sns.rugplot(df)
```

```python
sns.rugplot(df['tip'])
```
