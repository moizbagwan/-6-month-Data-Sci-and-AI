# Function in Python

    Python Functions is a block of statements that does a specific task. The idea is to put some commonly or repeatedly done task together and make a function so that instead of writing the same code again and again for different inputs, we can do the function calls to reuse code contained in it over and over again.

In Short:
- A function is a block of code that performs a specific task.
- Functions help us to reuse code instead of writing it again and again.
- A function runs only when we call it.

Benefits of Using Functions:
- Code Reuse
- Reduced code length
- Increased redability of code

Why use Functions?

    1- Code reusability → Write once, use many times.
    2- Code readability → Makes programs easy to understand.
    3- Easy debugging → Errors are easier to fix.
    4- Modularity → Large programs can be divided into small parts (functions).

# def keyword
    The def keyword stands for define. It is used to create a user-defined function. It marks the beginning of a function block and allows you to group a set of statements so they can be reused when the function is called.


Syntax:

    def function_name(parameters):
        # function body

# Explanation:

- def: Starts the function definition.
- function_name: Name of the function.
- parameters: Inputs passed to the function (inside ()), optional.
- : : Indicates the start of the function body.
- Indented code: The function body that runs when called.

# Formal & Actual Arguments:

(1) Formal Arguments (a, b in definition)

- Variables written inside the function definition.
- They are just placeholders (waiting for real values).

(2) Actual Arguments (5, 6 in function call)

- Real values given to the function when it is called.
- These values are passed to the formal arguments.

```python
def add(a, b):   # a & b are formal arguments (placeholders)
    c = a + b
    print(c)

add(5, 6)        # 5 & 6 are actual arguments (real values)
```

# Types of Actual Arguments:

    i Positional arguments
    ii Keyword argument
    iii Default argument 
    iv Variable length argument

i Positional arguments:

    We used the Position argument during the function call so that the first argument (or value) is assigned to name and the second argument (or value) is assigned to age. By changing the position, or if you forget the order of the positions, the values can be used in the wrong places

```python
def person(name,age):
    print(name)
    print(age)
    
person('nit', 22)
```

```python
def person(name, age):
    print("Name:", name)
    print("Age:", age)

person("Moiz", 22)   # "Moiz" → name, 22 → age
```

# ii Keyword argument:

    The idea is to allow the caller to specify the argument name with values so that the caller does not need to remember the order of parameters.

```python
def person(name, age):
    print("Name:", name)
    print("Age:", age)

person(age=22, name="Moiz")
```

```python
def person(name,age):
    print(name)
    print(age)
    
person(22,'nit')
```

# iii Default argument:

    A default argument is a parameter that assumes a default value if a value is not provided in the function call for that argument.

```python
def person(name, age=18):
    print("Name:", name)
    print("Age:", age)

person("Abdullah", 25)   # both values given
person("Moiz")      # only name given
```

```python
def person(name, age, phone):
    
    print(name)
    print(age+1)
    print(phone)
    
person(age = 22, name = 'Moiz', phone = 78418)
```

```python
def person(name, age=18):
    
    print(name)
    print(age)
    
person('nit')
```
