# Iteration by Index of Sequences

    - We can also use the index of elements in the sequence of iterate. The key idea is to first the calculte the length of the list and then iterate over the sequence within the range of this length.

```python
li = ["Good", "Morning"]
for index in range(len(li)):
    print(li[index])
```

Explanation: This code iterates through each element of the list using its index and prints each element one by one. The range(len(list)) generates indices from 0 to the length of the list minus 1

```python
a = 'Hello'
for index in range(len(a)):
    print(a[index])
```

# Using ELSE Statement with For Loop

    - We can also combine else statement with for loop like in while loop. But as there is no condition in for loop based on which the execution will terminate so the else block will be executed immediately after for block finishes execution.

```python
a = ['Hello World']
for index in range(len(a)):
    print(a[index])
    
else:
    print("Bye for Now")
```

Explanation: The code iterates through the list and prints each element. After the loop ends it prints "Bye for now" as the else block executes when the loop completes without a break.

```python
s = ['MOIZ BAGWAN']
for index in range(len(s)):
    print(s[index])

else:
    print('Thank You!')
```

# While Loop

    - Runs until a condition becomes false.
    - Useful when number of iterations is not fixed.

** syntax:

        while expression:
        statement(s)

```python
img = plt.imread(r"C:\Users\hp\Downloads\While-Loop.webp")
plt.imshow(img)
plt.axis('off')
```

```python
a = 0
while a < 3:
    a = a +1
    print('Hello')
```

This example demonstrates a basic while loop in Python. The loop runs as long as the condition cnt < 3 is true. It increments the counter by 1 on each iteration and prints "Hello" three times.

# Using ELSE statement with While Loop:

    The else block in a while loop runs only when the condition becomes false naturally.
    If you stop the loop early with break, or if an error happens, then the else block will not run.

** Syntax:

    while condition:
        # execute these statements
    else:
        # execute these statements

```python
a = 0 
while a < 3 :
    a = a + 1
    print("Hello")
    
else:
    print('Bye for Now')
```

```python
sns = 9
while sns > 6:
    sns = sns - 1
    print('system')
else:
    print('See you')
```

# Nested Loops

    Python programming language allows to use one loop inside another loop which is called nested loop.
    A nested while loop means having one while loop inside another while loop.
    The outer loop runs first, and for each step of it, the inner loop runs completely.

```python
img = plt.imread(r"C:\Users\hp\Downloads\Nested-While-Loop.webp")    # we can off the axis.
plt.imshow(img)
plt.axis('off')
```

```python
i = 1
while i <= 2:       # Outer loop
    j = 1
    while j <= 3:   # Inner loop
        print(i, j)
        j += 1
    i += 1
```

```python
a = 1
while a <= 3:       # Outer loop
    b = 1
    while b <= 4:   # Inner loop
        print(a,b,end='')
        b += 1
    a += 1
```

# Loop Control Statements

    continue  skips to the next loop iteration.
    break  exits the loop.
    pass  does nothing (used as a placeholder).

# Continue Statement:
    The continue statement is used inside loops (for or while) to skip the rest of the code in the current iteration and move directly to the next iteration of the loop.

```python
for letter in 'Abdul Moiz':
    if letter == 'e' or letter == 's':
        continue
    print("Current letter",letter)

```

Explanation: The continue statement is used to skip the current iteration of a loop and move to the next iteration. It is useful when we want to bypass certain conditions without terminating the loop.

```python
for i in range(1,6):
    if i == 3:
        continue # skip i = 3.
    print(i)
```

```python
for i in range(1,11):
    if i <= 4:
        continue   # skip i = 4.
    print(i)
```

```python
i = 0
while i < 5:
    i +=1
    if i == 2:
        continue     # skip when i = 2.
    print(i)

```

```python
j = 0
while j <10:
    j +=1
    if j == 8:    # skip when j = 8.
        continue
    print(j)
```

# Break Statement:

    - The break statement is used to terminate a loop immediately (whether its a for loop or a while loop).
    - Once Python encounters break, it exits the loop and continues executing the code after the loop.

```python
for i in range(1,10):
    if i == 5:
        break    # Exit the loop when i is 5
    print(i)
```

```python
for j in range(1,20):
    if j == 10:
        break      # Exit the loop when j is 45
    print(j,end='')
```

```python
i = 0
while i < 10:
    i += 1
    if i == 5:
        break
    print(i)
```

```python
j = 1
while j <= 10:
    if j == 6:
        break    # Exit the loop when n is 6
    print(j)
    j += 1
```

Explanation: break statement is used to exit the loop prematurely when a specified condition is met

# Pass Statement:

    - We use pass statement in Python to write empty loops. Pass is also used for empty control statements, functions and classes.
    
                                (OR)
                                
    - It is mainly used as a placeholder when you want to write code later but still need syntactically valid Python code.

```python
for i in range(5):
    pass               # does nothing for now.

#Here, the loop runs 5 times, but pass means "do nothing". It just keeps the loop body empty without causing an error.
```
