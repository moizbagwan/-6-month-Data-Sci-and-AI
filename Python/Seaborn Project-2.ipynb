- Also if you look up you can find the uniform distribution (Audience)

```python
j = sns.jointplot(data = movies, x = 'CriticRating', y = 'AudienceRating')
plt.show()

# Audience rating is more dominant then critics rating
# Based on this we find out as most people are most liklihood to watch audience rating & less likely to wathc critics rating
# let me explain the excel - if you filter audience rating & critic rating. critic rating has very low values compare to audience rating 
```

```python
# Python is not vectorize programing language
# Building dashbords (dashboard - combination of charts).

sns.set_style('darkgrid')
f, axes = plt.subplots (2,2, figsize = (15,5))

k1 = sns.kdeplot(data = movies, x = movies.BudgetMillions, y = movies.AudienceRating,ax = axes[0,0])
k2 = sns.kdeplot(data = movies, x = movies.BudgetMillions, y = movies.CriticRating, ax = axes[0,1])

k1.set(xlim=(-20,160))
k2.set(xlim=(-20,160))

z = sns.violinplot(data = movies[movies.Genre == 'Drama'], x = 'Year', y = 'CriticRating',ax=axes[1,0])
k4 = sns.kdeplot(data=movies,x = movies.CriticRating, y = movies.AudienceRating,shade = True,shade_lowest=False,cmap='Reds',ax=axes[1,1])
k4b = sns.kdeplot(data=movies, x = movies.CriticRating, y = movies.AudienceRating,cmap='Reds',ax=axes[1,1])
plt.show()
```

```python
# How can you style dashboard using different color map
# python is not vectorize programming language
# Building dashboards (dashboard - combination of chats)

sns.set_style('dark',{'axes.facecolor':'black'})
f,axes = plt.subplots(2,2, figsize = (15,15))

#plot[0,0]
k1 = sns.kdeplot(data = movies, x = movies.BudgetMillions, y = movies.AudienceRating, \
                shade = True, shade_lowest=True,cmap = 'inferno', \
                ax=axes[0,0])

k1b = sns.kdeplot(data = movies, x = movies.BudgetMillions, y = movies.AudienceRating,\
                 cmap = 'cool',ax=axes[0,0])

k2 = sns.kdeplot(data = movies, x = movies.BudgetMillions, y = movies.CriticRating,\
               shade = True,shade_lowest=True,cmap= 'inferno',\
               ax=axes[0,1])

k2b = sns.kdeplot(data=movies, x = movies.BudgetMillions, y = movies.CriticRating,\
                 cmap = 'cool',ax=axes[0,1])

# plot[1,0]
z = sns.violinplot(data=movies[movies.Genre=='Drama'],\
                  x = 'Year', y = 'CriticRating',ax=axes[1,0])

#plot[1,1]
k4 = sns.kdeplot(data=movies, x = movies.CriticRating, y = movies.AudienceRating, \
                 shade =True,shade_lowest=False,cmap='Blues_r', \
                 ax = axes[1,1])

k4b = sns.kdeplot(data=movies, x = movies.CriticRating, y = movies.AudienceRating, \
                 cmap = 'gist_gray_r', ax = axes[1,1])


k1.set(xlim=(-20,150))
k2.set(xlim=(-20,160))

plt.show()
```
