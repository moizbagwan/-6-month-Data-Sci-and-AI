# ***Exceptional Handling***

**What is Exception Handling?**

- Exception Handling is a way to handle errors (exceptions) in your program so that the program does not crash and can continue running or exit gracefully.
- An exception is an error that occurs during the execution of a program (runtime error).
- Common examples: division by zero, file not found, invalid input, etc.

**Keywords in Exception Handling**

*Python uses four main keywords:*

- try – Code that might raise an exception goes here.
- cept – Code that runs if an exception occurs.
- else – (Optional) Code that runs if no exception occurs.
- finally – (Optional) Code that always runs, no matter what.

**Key Points**

- Exception handling prevents program crashes.
- You can catch specific exceptions or all exceptions using except Exception as e.
- finally is often used to close files, release resources, or cleanup.

# ***What is an Error in Python?***

- An error is something that goes wrong in the program.
- Python gives feedback when an error occurs, which helps you debug it.
  
***Errors are mainly of Three types:***
  
*  1. *Syntax Error*
*  2. *Logical Error*
*  3. *Run Time Error*

# ***1. Syntax Error (Compile-time errors)***

**Definition**

- **A syntax error occurs when the programmer writes code that violates the rules (grammar) of the programming language.**
- The interpreter/compiler cannot understand the code, so it fails to run.
- These errors are detected before the program executes fully.

**Characteristics**

1. Detected before or at program execution.
2. The program cannot run until the syntax error is fixed.
3. Usually caused by typos or missing punctuation.
4. Python usually shows an error message with line number, making it easier to locate.


```python
# Common Causes of Syntax Errors
# 1. Missing colon :

if 5 > 2   # Missing colon
    print("Yes")

# Error: SyntaxError: expected ':'
```

```python
# 2. Incorrect indentation :

def hello():
print("Hello")  # Should be indented

# Error: IndentationError: expected an indented block
```

```python
# 3. Unmatched parentheses, brackets, or quotes :

print("Hello World)  # Closing quote missing

# Error: SyntaxError: EOL while scanning string literal
```

```python
# 4. Using keywords incorrectly :

for = 5  # 'for' is a keyword, cannot be variable name

# Error: SyntaxError: invalid syntax
```

```python
# 5. Wrong function definition :

def greet  # Missing parentheses
    print("Hi")

# Error: SyntaxError: expected '('


#------------------------------
# Syntax errors are mistakes in the structure of code, preventing it from running. Fixing them is the first step to making a program work.
#-----------------------------
```

**How to Fix Syntax Errors**

- Read the error message carefully. Python usually tells the line number.
- Check for typos: missing colons, parentheses, quotes.
- Check indentation: Python requires consistent spaces or tabs.
- Verify keyword usage: Don’t use reserved words as variable names.
- Test small blocks of code to isolate the problem.

```python
# Wrong Code (Syntax Error)

name = input("Enter your name"
print("Hello, " + name)

# eRROR : SyntaxError: '(' was never closed (unexpected EOF while parsing)
```

```python
#------------------------------------------
# Correct Code
#------------------------------------------

name = input("Enter your name")  # Closed parenthesis
print("Hello, " + name)
```

# ***2. Logical Error***

**Definition**

- **A logical error occurs when the program runs without crashing, but the output is not what you expected.**
- The code syntax is correct, so the interpreter/compiler does not show an error.
- The logic or reasoning in the code is wrong.
- These are the hardest errors to find because the program executes, but results are incorrect.

**Characteristics**

1. Program runs successfully (no syntax or runtime errors).
2. Output is incorrect or unexpected.
3. Caused by wrong formulas, conditions, loops, or assumptions in code.
4. Debugging requires careful checking of logic and testing with different inputs.


```python
# 1. Wrong arithmetic operations :

x = 10
y = 5
result = x - y   # Suppose we wanted x + y
print(result)    # Output: 5 (Incorrect for intended logic)
```

```python
# 2. Incorrect conditional statements :

age = 15
if age > 18:
    print("You can vote")
else:
    print("You cannot vote")

# Error: Logic is wrong if we wanted 15 to be eligible (depends on the rule).
```

```python
# 3. Wrong loop conditions :

for i in range(1, 10):
    print(i*2)  # If we wanted i*3, this is a logical error
```

```python
# 4. Incorrect use of functions or variables :

def add(a, b):
    return a - b  # Subtracting instead of adding

print(add(5, 3))  # Output: 2, should be 8
```

**How to Detect Logical Errors**

- Test with sample inputs and check outputs.
- Print intermediate values to trace calculation steps.
- Use a debugger to follow the program flow.
- Write unit tests to verify that functions give correct results.

```python
# 5. Incorrect Logic :
# -----------------------------------------------------------

# Program to find the average of two numbers
a = 10
b = 20
average = a + b / 2  # Incorrect logic: should use (a+b)/2
print("Average with INCORRECT LOGIC:", average)

#-------------------------------------------------------------
# Correct Logic
#-------------------------------------------------------------

average = (a + b) / 2
print("Average with CORRECT LOGIC:", average)
```

***Key Point:***

*- Logical errors don’t stop your program from running; they just make it behave incorrectly. Careful testing and debugging are the only ways to find them.*

# ***3. Runtime Errors (Execution-time errors)***

**Definition**

- A runtime error occurs while the program is running.
- The syntax is correct, so the program starts executing.
- Something happens during execution that prevents the program from continuing, and Python shows an error message.
- These are also called exceptions in Python.

**Characteristics**

1. Program starts running but crashes at some point.
2. Caused by invalid operations, wrong input, or unexpected situations.
3. Can be handled using Exception Handling in Python (try-except).
4. Examples include dividing by zero, accessing a missing file, or using an undefined variable.

| IndexError        | `lst = [1,2]; print(lst[5])` | Accessing an invalid index in a list     |
| FileNotFoundError | `open("file.txt")`           | File does not exist                      |
| ValueError        | `int("abc")`                 | Invalid value for a conversion           |

# ***What is try-except Block?***

***Definition:***
                **-" try-except is a way to try a block of code and handle errors if they occur, so the program can continue running."**
    
- In Python, some statements may cause errors when executed. These are called critical statements.
- If an error happens and we don’t handle it, the program stops immediately.
- To prevent the program from crashing, Python provides exception handling using try-except.

- try: → Python will attempt to run the code inside this block.
- except: → Python will execute this block only if an error occurs.
- You can also catch specific errors, e.g., ZeroDivisionError or FileNotFoundError.

# ***Finally Block***

**Definition:**

- **finally is a block of code that is executed after the try and except blocks, whether an exception occurred or not. It is often used to release resources like files, database connections, or network connections.**

**Key Points:**

- Always runs — even if there is an error or a return statement in try or except.
- Commonly used for cleanup tasks, like closing files or releasing resources. 
- Optional — you can use finally with or without except.

```python
try:
    a = 5
    b = 0
    
    result = a / b     # Critical statement
    print(result)
except ZeroDivisionError:
    print("Cannot divide by zero!")  # Handle the error

print("Bye")  # This will run no matter what
```

```python
c = 5        # assign 5 to variable c
d = 0        # assign 0 to variable d

try:
    print(c/d)   # critical statement: may cause ZeroDivisionError

except Exception:  # catch any exception that occurs in try block
    print("hey you can not devive Number by zero")  # handle the error

print('Bye')  # this will run regardless of error


#-------------------------------------------
# Note:
# Using Exception in except will catch all types of errors, not just division by zero.
#The print('Bye') runs even if an error occurs, showing the program didn’t crash.
```

```python
c = 3        # assign 3 to variable c
d = 0        # assign 0 to variable d

try:
    print(d/c)   # critical statement: divide d by c

except Exception:  # if any error occurs in try block
    print('hey you cannot divide Number by zero')  # handle the error gracefully

print('bye')  # this will run no matter what, even if there was an error
```

```python
c = 3        # assign 3 to variable c
d = 0        # assign 0 to variable d

try:
    print(c/d)   # critical statement: may cause ZeroDivisionError

except Exception as e:  # catch any exception and store details in variable e
    print('hey you cannot divide Number by zero:', e)  # print error message along with actual exception

print('bye')  # this will run regardless of whether an error occurred
```

```python
c = 3
d = 3

try:
    print('resource open')      # pretend we are opening a resource (like a file or database)
    print(c/d)                  # divide c by d (this may cause an error)

except Exception as e:           # if any error happens in try block
    print('hey you cannot divide Number by zero', ':', e)  # show error message
    print('Resource closed')     # close the resource if error occurs

finally:
    print('resource closed')     # this runs always, whether error occurs or not



# ------------Simple explanation:--------------------

# try → tries to run the code that might fail.
# except → handles the error if it happens.
# finally → always runs, used to ensure resources are closed even if there is an error.
# This is the safe way to work with critical statements like file/database operations.
```
