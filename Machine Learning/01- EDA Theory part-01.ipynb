# 1) Variable Identification
                            It is the First step to analys dataset, Variable Identification means figuring out which columns in your data are features (input) and which is the target (output), and what type they are. 

# Types of Variable Identification:

    - there are Two (2) types are as follows:

    1) Independent Variable --> Used to predict
    2) Dependent Variable  ---> What we want to predict


# i) Independent Variable
    Independent Variable = The variable we use to predict something else.

Key Points:

    Independent variable = input / predictor
    Can be numerical or categorical
    One dataset can have many independent variables

# ii) Dependent Variable

    Dependent variable is the output we want to predict or explain in a dataset.
    - It depends on independent variables
    - Also called target variable



| **Type**                                       | **Meaning**                              | **Example**                                        |
| ---------------------------------------------- | ---------------------------------------- | -------------------------------------------------- |
| **Independent Variable (Feature / Predictor)** | Input variable used to predict something | `Age`, `BMI`, `Smoker` (used to predict `Charges`) |
| **Dependent Variable (Target / Outcome)**      | Output variable we want to predict       | `Charges` (depends on `Age`, `BMI`, `Smoker`)      |

# 2) Univariate analysis:
                Univariate analysis is analyzing one variable at a time to understand its distribution, summary statistics, and patterns.

                                                                        [OR]
                                                                        
                Univariate Analysis = Looking at one variable at a time to understand its values, distribution, and patterns.
- “Uni” = one variable
- Focus is on single column/feature

Purpose:
- Check distribution (how data is spread)
- Find mean, median, mode, min, max
- Detect outliers
- Understand data type

-------------------------------------------
# 3) Bivariate Analysis
                Bivariate analysis means analyzing the relationship between two variables.

- “Bi” = two variables
- Helps us see how one variable affects or relates to another

Purpose

- Understand correlation or association between variables
- Find patterns or trends
- Prepare for modeling

-----------------------

# 4) Outliers Treatments
        Outliers are values that are very different from most of the data.
        Treatment means handling or fixing them so they don’t spoil analysis or models.

Why Treat Outliers?
- Can distort mean, standard deviation, correlation
- Can affect machine learning models.
- Sometimes they are important (like fraud detection), so treatment depends on context.

Common Methods to Treat Outliers:
- Remove Outliers
- Remove data points outside a range
- Example: using IQR (Interquartile Range) method

-------------------------------

# 5) Missing Value Treatments
        Missing values are empty or null entries in the dataset.
        Treatment means handling them so analysis or modeling is correct.

Why Treat Missing Values?
- Missing values can cause errors in analysis or models
- Can bias results if not handled properly


Fill / Impute Missing Values
Replace missing data with a value

- For Numerical Columns: use (mean, median, mode startagy)

- Mean: df['age'].fillna(df['age'].mean(), inplace=True)

- Median: df['age'].fillna(df['age'].median(), inplace=True)

- Mode: df['age'].fillna(df['age'].mode()[0], inplace=True)

- For Categorical Columns:(mode startagy)
- Mode (most frequent value)

df['gender'].fillna(df['gender'].mode()[0], inplace=True)

# 4) Imputation / Transformer

Imputation:
            - Imputation means filling or replacing missing values in a dataset so that analysis or modeling can be done properly.

- Missing values can cause errors or bias in analysis
- Imputation estimates a value for missing data

Transformer:
            - Transformer = Change or scale your data to make it suitable for analysis or modeling.
            - Helps models learn better and improves performance
