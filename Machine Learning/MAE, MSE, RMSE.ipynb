```python
import numpy as np
import matplotlib.pyplot as plt

# Create a range of error values from -10 to 10 (200 points)
errors = np.linspace(-10, 10, 200)

# Calculate Mean Absolute Error (MAE) for each error value
# MAE = |error|
mae = np.abs(errors)

# Create a figure for plotting with specific size
plt.figure(figsize=(6, 4))

# Plot the error values (x-axis) against their MAE (y-axis)
plt.plot(errors, mae, label="MAE = |error|", color="blue", linewidth=2)

# Add title and axis labels
plt.title("MAE Graph")
plt.xlabel("Error value")
plt.ylabel("MAE")

# Add grid lines for better readability
plt.grid(True)

# Show the legend to explain the line meaning
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Actual and predicted values
actual = [3, 5, 2, 7]
predicted = [2, 5, 4, 8]

# Calculate the difference (error) between actual and predicted values
errors = [a - p for a, p in zip(actual, predicted)]

# Square each error to get the squared error (used for MSE calculation)
squared_errors = [e**2 for e in errors]  # MSE for each individual data point

# Create a range for the x-axis (1 to number of data points)
x = range(1, len(actual) + 1)

# Plot the squared errors as a line graph
plt.plot(x, squared_errors, marker='o', color='green', linestyle='-')

# Label the x-axis and y-axis
plt.xlabel('Data Point')
plt.ylabel('Squared Error (MSE)')

# Add a title to the plot
plt.title('MSE Line Graph')

# Display a grid for better visualization
plt.grid(True)

# Show the final plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Actual and predicted values
actual = [3, 5, 2, 7]
predicted = [2, 5, 4, 8]

# Calculate errors
errors = [a - p for a, p in zip(actual, predicted)]

# Calculate MAE (Mean Absolute Error)
abs_errors = [abs(e) for e in errors]          # MAE

# Calculate MSE (Mean Squared Error)
squared_errors = [e**2 for e in errors]        # MSE

# Calculate RMSE (Root Mean Squared Error)
rmse = [e**0.5 for e in squared_errors]        # RMSE

# X-axis values for data points
x = range(1, len(actual) + 1)

# Create 3 vertical subplots with increased height for clarity
fig, axs = plt.subplots(3, 1, figsize=(8, 10))  # Height increased from 1 to 10

# MAE subplot as bar chart
axs[0].bar(x, abs_errors, color='skyblue')
axs[0].set_title('MAE (Mean Absolute Error)')
axs[0].set_ylabel('Absolute Error')
axs[0].set_xticks(x)
axs[0].grid(True, linestyle='--', alpha=0.5)

# MSE subplot as bar chart
axs[1].bar(x, squared_errors, color='lightgreen')
axs[1].set_title('MSE (Mean Squared Error)')
axs[1].set_ylabel('Squared Error')
axs[1].set_xticks(x)
axs[1].grid(True, linestyle='--', alpha=0.5)

# RMSE subplot as bar chart
axs[2].bar(x, rmse, color='salmon')
axs[2].set_title('RMSE (Root Mean Squared Error)')
axs[2].set_ylabel('Error (Original Unit)')
axs[2].set_xlabel('Data Point')
axs[2].set_xticks(x)
axs[2].grid(True, linestyle='--', alpha=0.5)

# Adjust layout so titles, labels, and bars do not overlap
plt.tight_layout()

# Display the plot
plt.show()
```

| Metric   | Full Form               | Formula                                            | Example (Actual=\[3,5,2,7], Predicted=\[2,5,4,8])     | Key Points                                          |                                             |                                                                 |
| -------- | ----------------------- | -------------------------------------------------- | ----------------------------------------------------- | --------------------------------------------------- | ------------------------------------------- | --------------------------------------------------------------- |
| **MAE**  | Mean Absolute Error     | ( \text{MAE} = \frac{1}{n}\sum                     | y\_i - \hat{y}\_i                                     | )                                                   | Errors = \[1,0,2,1] → MAE = (1+0+2+1)/4 = 1 | Simple, treats all errors equally, less sensitive to big errors |
| **MSE**  | Mean Squared Error      | $\text{MSE} = \frac{1}{n}\sum (y_i - \hat{y}_i)^2$ | Squared Errors = \[1,0,4,1] → MSE = (1+0+4+1)/4 = 1.5 | Penalizes large errors more, unit is squared        |                                             |                                                                 |
| **RMSE** | Root Mean Squared Error | $\text{RMSE} = \sqrt{\text{MSE}}$                  | RMSE = √1.5 ≈ 1.22                                    | Same unit as original data, sensitive to big errors |                                             |                                                                 |


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Population parameters
mu = 60        # Population mean
sigma = 5      # Population standard deviation
n = 40         # Sample size

# Sample mean
x_bar = 62

# Calculate Z-score for the sample mean
# Z = (sample mean - population mean) / (population standard deviation / sqrt(sample size))
z = (x_bar - mu) / (sigma / np.sqrt(n))
print("Z-score:", z)

# X-axis values for plotting the population normal distribution
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
# Probability density function for normal distribution
y = norm.pdf(x, mu, sigma)

# Plot the population normal distribution
plt.plot(x, y, label='Population Distribution')
# Fill under the curve lightly for visual emphasis
plt.fill_between(x, 0, y, alpha=0.1)

# Draw a vertical line for the population mean
plt.axvline(mu, color='blue', linestyle='--', label='Population Mean (μ=60)')

# Draw a vertical line for the sample mean
plt.axvline(x_bar, color='red', linestyle='-', label=f'Sample Mean (x̄={x_bar})')

# Annotate the Z-score near the sample mean
plt.text(x_bar + 0.5, max(y)/2, f'Z = {z:.2f}', color='red')

# Add title and axis labels
plt.title('Z-Test Visualization')
plt.xlabel('Value')
plt.ylabel('Probability Density')

# Add legend and grid
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
```

```python
from scipy.stats import norm

# Z-score from previous example
z = 2.53

# Calculate two-tailed p-value
# Two-tailed because we are testing if the sample mean is significantly different (either direction)
p_value = 2 * (1 - norm.cdf(abs(z)))  # norm.cdf gives cumulative probability up to z
print("P-value:", p_value)

# Significance level
alpha = 0.05  # Commonly used threshold (5%)

# Hypothesis testing decision
if p_value < alpha:
    print("Reject H0 → Sample mean is significantly different from population mean")
else:
    print("Fail to reject H0 → Sample mean is NOT significantly different")
```
