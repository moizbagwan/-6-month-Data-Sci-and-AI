```python
# Select independent variables (features) for regression
# df.iloc[:, :-1] selects all rows (:) and all columns except the last one (:-1)
# This assumes the last column is the target variable
x = df.iloc[:, :-1]

# Select the dependent variable (target) for regression
# df.iloc[:, 4] selects all rows (:) and the 5th column (index 4)
# Make sure the column index matches your target variable
y = df.iloc[:, 4]

# Print the features (X) to verify selection
print(x)
```

```python
# Print the target variable (y) to verify its selection
# This helps ensure you have chosen the correct column for prediction
print(y)
```

```python
# Convert categorical variables in features (X) into dummy/indicator variables
# pd.get_dummies() automatically creates binary columns for each category
# dtype=int ensures the new columns are of integer type (0 or 1)
# This is necessary because regression models require numeric input
x = pd.get_dummies(x, dtype=int)

# Print the transformed features to verify that categorical columns are converted
print(x)
```

```python
# Split the dataset into training and testing sets
# x: independent variables (features)
# y: dependent variable (target)
# test_size=0.2 means 20% of the data will be used for testing, 80% for training
# random_state=0 ensures reproducibility of the split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Print the split datasets to verify
# x_train, y_train → training data for model learning
# x_test, y_test → testing data to evaluate model performance
print(x_train, x_test, y_train, y_test)

print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
# Create an instance of the LinearRegression model
regressor = LinearRegression()

# Train the model on the training data
# x_train → features for training
# y_train → corresponding target values
# The model learns the relationship between X and y (fits the best line)
regressor.fit(x_train, y_train)
```

```python
# Use the trained Linear Regression model to make predictions on the test data
# x_test → features for testing
# y_pred → predicted target values corresponding to x_test
y_pred = regressor.predict(x_test)

# Display the predicted values
y_pred
```

```python
# Get the coefficients (weights) of the trained Linear Regression model
# Each value in regressor.coef_ corresponds to the effect of one feature on the target
# A positive coefficient means the feature increases the target, negative means it decreases
m = regressor.coef_

# Print the coefficients to see how each feature impacts the prediction
print(m)
```

```python
# Get the intercept (bias) of the trained Linear Regression model
# The intercept is the predicted value of y when all features are 0
c = regressor.intercept_

# Print the intercept
print(c)
```

```python
# Add a column of constant values (42467) to the features X
# np.full((50,1), 42467) creates a 50x1 array filled with 42467
# .astype(int) ensures the values are integers
# np.append(..., values=x, axis=1) appends this new column to the existing X matrix along columns
# This could be used as a manual bias/intercept column for models like statsmodels OLS
x = np.append(arr=np.full((50,1), 42467).astype(int), values=x, axis=1)

# Display the updated feature matrix
print(x)
```

```python
# Select a subset of features (columns) for Ordinary Least Squares (OLS) regression
# Here, x[:, [0,1,2,3,4,5]] selects columns 0 to 5 from the feature matrix
# This is often done for feature selection or backward elimination
x_opt = x[:, [0,1,2,3,4,5]]

# Fit an OLS (Ordinary Least Squares) regression model using statsmodels
# endog=y → dependent variable (target)
# exog=x_opt → independent variables (features)
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()

# Display a detailed summary of the regression results
# The summary includes coefficients, standard errors, t-values, p-values, R-squared, etc.
regressor_OLS.summary()
```

```python
# Select a subset of features (columns) for OLS regression
# Here, columns 0,1,2,3,5 are selected (column 4 is removed)
# This is commonly done in backward elimination to remove statistically insignificant features
x_opt = x[:, [0,1,2,3,5]]

# Fit an Ordinary Least Squares (OLS) regression model using statsmodels
# endog=y → dependent variable (target)
# exog=x_opt → independent variables (selected features)
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()

# Display a detailed summary of the regression results
# The summary includes:
# - Coefficients for each feature
# - Standard errors
# - t-values and p-values (to check statistical significance)
# - R-squared and adjusted R-squared (to measure model fit)
regressor_OLS.summary()
```

```python
# Select a further reduced subset of features for OLS regression
# Here, columns 0,1,2,3 are selected (columns 4 and 5 removed)
# This is likely part of backward elimination to remove features with high p-values
x_opt = x[:, [0,1,2,3]]

# Fit an Ordinary Least Squares (OLS) regression model using statsmodels
# endog=y → dependent variable (target)
# exog=x_opt → independent variables (selected features)
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()

# Display a detailed summary of the regression results
# The summary includes:
# - Coefficients for each feature
# - Standard errors
# - t-values and p-values (to check statistical significance)
# - R-squared and adjusted R-squared (to measure model fit)
regressor_OLS.summary()
```

```python
# Select an even smaller subset of features for OLS regression
# Here, only columns 0, 1, and 2 are selected
# This is likely part of backward elimination to remove statistically insignificant features
x_opt = x[:, [0, 1, 2]]

# Fit an Ordinary Least Squares (OLS) regression model using statsmodels
# endog=y → dependent variable (target)
# exog=x_opt → independent variables (selected features)
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()

# Display a detailed summary of the regression results
# The summary provides:
# - Coefficients for each feature
# - Standard errors
# - t-values and p-values (to check statistical significance)
# - R-squared and adjusted R-squared (to evaluate model fit)
regressor_OLS.summary()
```

```python
# Select the final subset of features for OLS regression
# Here, only columns 0 and 1 are selected
# This is likely the last step in backward elimination, keeping only statistically significant features
x_opt = x[:, [0, 1]]

# Fit an Ordinary Least Squares (OLS) regression model using statsmodels
# endog=y → dependent variable (target)
# exog=x_opt → independent variables (selected features)
regressor_OLS = sm.OLS(endog=y, exog=x_opt).fit()

# Display a detailed summary of the regression results
# The summary includes:
# - Coefficients for each feature
# - Standard errors
# - t-values and p-values to check statistical significance
# - R-squared and adjusted R-squared to evaluate model fit
regressor_OLS.summary()
```

```python
# Calculate the R² score (coefficient of determination) of the trained Linear Regression model
# regressor.score() returns how well the model fits the training data
# x_train → training features
# y_train → training target values
# The R² score ranges from 0 to 1 (closer to 1 means better fit)
bias = regressor.score(x_train, y_train)

# Display the R² score of the model on training data
print(bias)
```

```python
# Calculate the R² score (coefficient of determination) of the trained Linear Regression model on test data
# x_test → testing features
# y_test → actual target values for testing
# This measures how well the model generalizes to unseen data
# The R² score ranges from 0 to 1 (closer to 1 means better prediction performance)
variance = regressor.score(x_test, y_test)

# Display the R² score of the model on test data
print(variance)
```

Multiple Linear Regression Workflow Summary:

1. Import Libraries
- NumPy & Pandas → handle numbers and datasets
- Matplotlib & Seaborn → visualize data
- Scikit-learn → ML models & train-test split
- Statsmodels → statistical modeling (OLS)

2. Load & Explore Data
- Load CSV: df = pd.read_csv(...)
- Inspect data: df.head(), df.info(), df.describe()
- Check for missing values: df.isnull().sum()
- Check columns & size: df.columns, df.shape

3. Prepare Features & Target
- Features: x = df.iloc[:, :-1]
- Target: y = df.iloc[:, 4]
- Encode categorical variables: pd.get_dummies(x)
- Optional: add constant/bias column manually with np.append(...)

4. Split Dataset
- train_test_split(x, y, test_size=0.2, random_state=0) → train/test sets

5. Train Linear Regression Model
- Create model: regressor = LinearRegression()
- Train: regressor.fit(x_train, y_train)

6. Predict & Inspect
- Predict: y_pred = regressor.predict(x_test)
- Model coefficients: regressor.coef_
- Intercept (bias): regressor.intercept_

7. Evaluate Model
- R² on training data: regressor.score(x_train, y_train)
- R² on test data: regressor.score(x_test, y_test)

8. Statsmodels OLS (Optional/Advanced)
- Run OLS regression: sm.OLS(endog=y, exog=x_opt).fit()
- Check summary: .summary() → coefficients, p-values, R²
- Backward elimination: remove features with high p-values iterativel

9. Key Takeaway:
- Your code uses multiple independent variables to predict a target → this is Multiple Linear Regression, which is a type of linear regression.
